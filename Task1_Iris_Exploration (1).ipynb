# Task 1: Exploring and Visualizing the Iris Dataset

## Introduction & Problem Statement
The goal of this task is to understand how to load, inspect, summarize, and visualize a real-world dataset. We use the classic **Iris Dataset**, which contains measurements of 150 iris flowers from three species: *Setosa*, *Versicolor*, and *Virginica*.

### Objectives:
- Load and inspect the dataset structure
- Summarize key statistics
- Create scatter plots, histograms, and box plots to understand the data

```python
# ── Install / import libraries ──────────────────────────────────────────────
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print('Libraries loaded successfully!')
```

## 1. Dataset Understanding & Description
The Iris dataset has 150 rows and 5 columns:
- `sepal_length`, `sepal_width`, `petal_length`, `petal_width` — numerical measurements in cm
- `species` — categorical target with 3 classes

```python
# ── Load dataset via seaborn (no file download needed) ──────────────────────
df = sns.load_dataset('iris')

print('=== Dataset Shape ===')
print(df.shape)           # rows × columns

print('\n=== Column Names ===')
print(df.columns.tolist())

print('\n=== First 5 Rows ===')
df.head()
```

```python
# ── Data types & missing values ─────────────────────────────────────────────
print('=== Data Types ===')
print(df.dtypes)

print('\n=== Missing Values ===')
print(df.isnull().sum())
```

```python
# ── Statistical summary ─────────────────────────────────────────────────────
print('=== Statistical Summary ===')
df.describe()
```

```python
# ── Class distribution ──────────────────────────────────────────────────────
print('=== Species Count ===')
print(df['species'].value_counts())
```

## 2. Exploratory Data Analysis (EDA) with Graphs

```python
# ── Scatter Plot: Sepal Length vs Sepal Width ────────────────────────────────
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width',
                hue='species', palette='Set1', s=80)
plt.title('Scatter Plot: Sepal Length vs Sepal Width', fontsize=14)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
print('Insight: Setosa has smaller sepal length but larger sepal width compared to other species.')
```

```python
# ── Scatter Plot: Petal Length vs Petal Width ────────────────────────────────
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width',
                hue='species', palette='Set2', s=80)
plt.title('Scatter Plot: Petal Length vs Petal Width', fontsize=14)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
print('Insight: Petal measurements clearly separate all three species — great features for classification.')
```

```python
# ── Histograms for all numeric features ──────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Histograms of Iris Features by Species', fontsize=15)

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for ax, feature in zip(axes.flatten(), features):
    for species in df['species'].unique():
        ax.hist(df[df['species'] == species][feature], bins=15,
                alpha=0.6, label=species)
    ax.set_title(feature.replace('_', ' ').title())
    ax.set_xlabel('Value (cm)')
    ax.set_ylabel('Frequency')
    ax.legend()

plt.tight_layout()
plt.show()
print('Insight: Petal length and petal width show the most distinct separation between species.')
```

```python
# ── Box Plots to detect outliers ─────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Box Plots of Iris Features by Species', fontsize=15)

for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(data=df, x='species', y=feature, palette='pastel', ax=ax)
    ax.set_title(feature.replace('_', ' ').title())
    ax.set_xlabel('Species')
    ax.set_ylabel('Value (cm)')

plt.tight_layout()
plt.show()
print('Insight: A few outliers exist in sepal_width for Virginica and Versicolor, but overall data is clean.')
```

```python
# ── Pairplot — relationships between all features ─────────────────────────────
sns.pairplot(df, hue='species', palette='Set1', diag_kind='kde')
plt.suptitle('Pairplot of All Iris Features', y=1.02, fontsize=14)
plt.show()
print('Insight: Petal length and petal width together provide the best visual separation of species.')
```

```python
# ── Correlation Heatmap ───────────────────────────────────────────────────────
plt.figure(figsize=(7, 5))
corr = df.drop(columns='species').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()
print('Insight: Petal length and petal width are strongly correlated (r=0.96).')
```

## 3. Conclusion

- The Iris dataset is **clean** — no missing values, 150 balanced samples.
- **Petal length** and **petal width** are the most discriminating features; they show clear clusters per species.
- **Setosa** is easily separable from the other two species; Versicolor and Virginica overlap slightly in sepal measurements.
- Correlation analysis shows petal features are highly correlated (0.96), meaning they carry similar information.
- This dataset is an excellent candidate for classification models (e.g., KNN, Decision Tree, Logistic Regression).
